This program shows: (1) how to plot a straight line, (2) how to translate that straight line to any angle, and (3) how to create a function that will plot a series of lines and various angles.

We will use these simple geometric scripts when we look at light reflecting off of surfaces.

```python
# Module Library used for this exercise

import matplotlib.pyplot as plt
import numpy as np
import math
%matplotlib inline
```

```python
# Establish the arrays for a straight line

i_y = np.arange(0,10)
i_x =np.array([0.0]*len(i_y))
```

```python
# Plotting the straight line

plt.plot(i_x, i_y)
plt.gca().set_aspect('equal', adjustable='box') 
# the above function creates a square frame for the plot which provides a better representation than the default rectangular frame
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.show()
```

```python
# translate the straight line to any given angle
# rotation matrix; x' = xcos(i_angle) + ysin(i_angle); y' = ycos(90-i_angle) - xsin(9-i_angle) for x, y

i_angle = float(raw_input('input the angle: '))
i_y_1 = i_y*math.cos(math.radians(i_angle))  
i_x_1 = i_x*math.cos(math.radians(i_angle)) + i_y*math.sin(math.radians(i_angle))                                                  
```

```python
# Plotting the translated line, and a reflected line

plt.plot(i_x_1, i_y_1, label = 'incident line')
plt.plot(-i_x_1, i_y_1, alpha = 0.5, label = 'reflected line')
plt.gca().set_aspect('equal', adjustable='box')  
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.legend(loc = 3)
plt.show()
```

```python
# A function to calculate and plot a series of tranlated lines

i_angle = np.arange(0, i_angle, 2)
for i in i_angle:
    i_y_1 = i_y*math.cos(math.radians(i)) 
    i_x_1 = i_x*math.cos(math.radians(i)) + i_y*math.sin(math.radians(i))
    plt.plot(i_x_1, i_y_1)
plt.gca().set_aspect('equal', adjustable='box')
plt.xlim(-10, 10)
plt.ylim(-10, 10) 
```
