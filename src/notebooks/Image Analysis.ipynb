```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.misc import imread
%matplotlib inline
```

```python
# img = imread('C:/Users/Timothy/Dropbox/File _requests/Hawaii_Astronomy_Conf_Presentation/Bob Adams - Waikola - Surface.png').astype(np.float32)/255
img = imread('C:/Users/Timothy/Desktop/northern-lights-picture-aurora-borealis-september-2011-yukon_41173_600x450.jpg').astype(np.float32)/255
img.shape
```

```python
def rgbw(img):
    red, green, blue = np.zeros(shape=(img.shape[0], img.shape[1], 3)), np.zeros(shape=(img.shape[0], img.shape[1], 3)), np.zeros(shape=(img.shape[0], img.shape[1], 3))
    red[:,:,0], green[:,:,1], blue[:,:,2] = img[:,:,0], img[:,:,1], img[:,:,2]
    blue_, green_, red_ = pd.DataFrame(blue[:,:,2]), pd.DataFrame(green[:,:,1]),pd.DataFrame(red[:,:,0])
    return red_, green_ , blue_, red, green, blue

red_, green_, blue_, red, green, blue = rgbw(img)
```

```python
plt.imshow(red), plt.axis('off'), plt.show()
plt.imshow(green), plt.axis('off'), plt.show()
plt.imshow(blue), plt.axis('off'), plt.show()
plt.imshow(img), plt.axis('off'), plt.show()
```

```python
Xb, Xg, Xr = blue_.columns.values, green_.columns.values, red_.columns.values
Yb, Yg, Yr = blue_.index.values, green_.index.values, red_.index.values
Zb, Zg, Zr = blue_.values, green_.values, red_.values

_, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
ax1.contourf(Xb, Yb, Zb, 10, alpha=0.7, cmap=plt.cm.Blues)
ax1.invert_yaxis(), ax1.axis('off'), ax1.set_aspect(img.shape[0]/img.shape[0])
ax2.contourf(Xg, Yg, Zg, 10, alpha=0.7, aspect = img.shape[1]/img.shape[0], cmap=plt.cm.Greens)
ax2.invert_yaxis(), ax2.axis('off'), ax2.set_aspect(img.shape[0]/img.shape[0])
ax3.contourf(Xr, Yr, Zr, 10, alpha=0.7, aspect = img.shape[1]/img.shape[0], cmap=plt.cm.Reds)
ax3.invert_yaxis(), ax3.axis('off'), ax3.set_aspect(img.shape[0]/img.shape[0])
ax4.imshow(img)
ax4.axis('off')

plt.show()
```
