```python
from scipy.misc import imread
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
raw_image = imread("../lamplight/images/pano_subset.jpeg")
image     = np.array(raw_image, dtype=np.float)/255
plt.imshow(image)
_ = plt.axis('off')
```

```python
red = np.zeros(shape=image.shape)
red[:,:,0] = image[:,:,0]
plt.imshow(red)
_ = plt.axis('off')
```

```python
green = np.zeros(shape=image.shape)
green[:,:,1] = image[:,:,1]
plt.imshow(green)
_ = plt.axis('off')
```
